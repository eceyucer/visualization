# Data Visualization: Customizing Our Plots

Since last lecture on April 24th did not have coding component, I had no commits, per the instructions of Eshta at the beginning of the class.

```python
# load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

# sample data
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

```python
# making a line plot with bot y variable on the same axes
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1)
ax.plot(x,y2)
fig.show() # IDE dependent, not necessary for VS Code or Anaconda usually
```

## Adding a legend

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc='lower right')
fig.show()
```

### Modifying our legend

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right',
        frameon = True, # add frame around the legend
        fontsize = 12, # change font size
        ncol = 2, # specify number of columns
        shadow = True )
fig.show()
```

### Moving the legend outside of the plot area

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")

ax.legend(loc='upper left',
        bbox_to_anchor =(1, 1)) # specifies where we want to anchor the legend location on the figure
fig.show()
```

## Text and Annotations

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.text(10, 95, "This value is important!") # point defined is the beginning of the text
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.text(10, 95, "This value is important!",
        ha='center',    # alignment of text
        color = 'red',  # modify font colour
        size = 20)      # modify font size

fig.show()
```

```python
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10]) # define axes ranges for the ticks

ax.text(1, 5, ". Data:(1, 5)", transform=ax.transData) # with reference to data
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform=ax.transAxes) # with reference to axes object 0.5 of the way in x, 0.1 in y
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform=fig.transFigure) # with reference to figure 20% of the way up and right
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),
            arrowprops=dict(facecolor='black', shrink=0.05))

fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!',
            xy=(10, 95), xytext=(20, 94),
            arrowprops = dict(arrowstyle = "wedge",
                            color = "hotpink"))

fig.show()
```

## Axis Labels

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.yaxis.set_major_locator(plt.NullLocator()) # remove ticks and labels
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.xaxis.set_major_formatter(plt.NullFormatter()) # remove only labels
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.xaxis.set_major_locator(plt.MaxNLocator(3)) #specifying intervals for tick marks
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

plt.xticks(rotation=45, ha='right') # rotate tick marks, can have longer labels/names
```

## Activity: Modifying axis title fonts

```python
# specify serif font with indigo color
new_font = {'family':'serif', 'color':'indigo'} 

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')

plt.xlabel('Shiny New X Axis!', fontsize = 18, fontdict = new_font) # new x-label
```
