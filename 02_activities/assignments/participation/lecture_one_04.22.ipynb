### Intro <a class="anchor" id="intro"></a>

```python
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# intro to how to plot with matplotlib
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100,50)

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y)
```

```python
# bar graph
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y)
```

```python
# line graph
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
```

```python
# histogram
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
```

```python
# setting axis titles 
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
```

```python
# you can import fonts and utilize them in your graphs
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}
```

```python
# how to set font in labels
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
# moving the graph title location to left, default is middle
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

### Markers
[Different markers](https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers)

```python
# using different markers in plots
fig, ax = plt.subplots(figsize=(5, 3))

ax.scatter( x,
            y,
            marker='*',        
            color = "indigo") 

fig.show()
```

```python
# can combine linestyles and markers
fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(x,
        y,
        marker='*', 
        color = 'indigo', 
        linestyle = '--', 
        linewidth = 2)
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))

# changing the color of markers
ax.plot(x,y,marker='*', color = '#7425b9', linestyle = '--', linewidth = 2)
fig.show()
```

```python
# setting custom colors for marker, lines and grids
fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(x,y,marker='*',
        markersize = 12,              
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359',  
        markerfacecolor = '#000000')

ax.grid()
fig.show()
```

```python
# setting custom colors for marker, lines and grids
fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(x,y,marker='*',
        markersize = 12,              
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359',  
        markerfacecolor = '#000000')

ax.grid(color = "blue", linewidth = 0.5)
fig.show()
```

```python
# setting custom colors and styles for the grids
fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(x,y,marker='*',
        markersize = 12,              
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359',  
        markerfacecolor = '#000000')

ax.grid(axis='y', color='blue', linewidth=2, linestyle='-.')
fig.show()
```
